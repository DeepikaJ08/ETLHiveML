# Hierarchical Clustering/ Agglomerative Clustering

### Hierarchical Clustering is called as Agglomerative Clustering in Python
#### Also called Bottom-Up Clustering



```python
from warnings import filterwarnings
filterwarnings ('ignore')
```

## Step 1: Read the dataset

```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

## Step 2: Perform Basic Data Quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m[m>0]
```

```python
df.duplicated().sum()
```

### Problem statement : Segment the customers on basis of Age, Income and Spending

## Step 3: Separate X from df

### Unsupervised data doesnt have Target feature so no Y and only X

```python
X = df[['Age', 'Income', 'Spending']]
X.head()
```

## Step 4: Create Preprocessing Pipeline for X

#### X has only continuous features so no need to separate cat and con features
#### only num_pipe needed here

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

## Step 5: Find Optimal no. of clusters for given data

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
```

```python
n = list(range(2, 15))
n
```

```python
sil = []
for i in n:
    model = AgglomerativeClustering(n_clusters=i)
    clusters = model.fit_predict(X_pre)
    s = silhouette_score(X_pre, clusters)
    sil.append(s)
```

```python
sil
```

### 5(i) Plot Silhoutte scores

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(n, sil) # x and y axis
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouttee Scores')
plt.title('Silhoutte Scores for Different Clusters')
plt.grid()
plt.show()
```

### Conclusion: From above plot 'i' has the highest silhoutte score at n=6

## Step 6: Build the final model

```python
agc_model = AgglomerativeClustering(n_clusters=6)
clusters = agc_model.fit_predict(X_pre)
clusters
```

## Step 7: Save the above in X

```python
X['Cluster'] = clusters
X.head()
```

## Step 8: Centroids

### 8(i)Get Centroids fo each cluster

```python
cent = X.groupby(by='Cluster').mean()
cent
```

### 8(ii) Segmenting columns by clustering

```python
# get the labels of the clusters assigned to the data given
X['Cluster'].unique()
```

```python
for i in range(6):
    print(f'Cluster Number : {i}\n')
    display(X[X['Cluster']==i])
    print('\n----------------------------------------------------------------------\n')
```
