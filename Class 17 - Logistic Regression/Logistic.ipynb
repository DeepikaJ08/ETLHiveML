### Target Feature (Y) is categorical feature

#### Sigmoid Curves gives probability of prediction



```python
from warnings import filterwarnings
filterwarnings ('ignore')
```

## Step 1: Read the Train Dataset

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

## Step 2: Perform Basic Data Quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m[m>0]
```

```python
df.duplicated().sum()
```

## Step 3: Separate X and Y (species)

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
# Bar chart to count each species class

Y.value_counts().plot(kind='bar')
```

## Step 4: Apply Preprocessing for X

```python
X.dtypes
```

```python
from sklearn.pipeline import Pipeline
# though there are no missing values, we still import 'imputer' because the Sample data may have missing values
from sklearn.impute import SimpleImputer
# since X features are only numerical in this dataset, ONLY numerical pipeline is required 
# DONT REQUIRE Categorical pipeline & ColumnTransformer
from sklearn.preprocessing import StandardScaler 
```

```python
# ONLY numerical pipeline created as X has only numerical features in this dataset
# for this reason output is transformed to pandas in this step itself
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
# as there is only numerical pipeline, X is tranformed for this
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

## Step 5: Apply Train-Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

## Step 6: Build the Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

## Step 7: Evaluate the Model

### 7(i): Accuracy in Train & Test data

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### 7(ii): Predict Results for Train & Test data

```python
ypred_train = model.predict(xtrain)
ypred_test = model.predict(xtest)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

### 7(iii): Probabilities of Prediction

```python
yprob_train = model.predict_proba(xtrain)
yprob_test = model.predict_proba(xtest)
```

```python
model.classes_
```

```python
yprob_train[0:5]
```

```python
yprob_test[0:5]
```

### 7(iv): Plot Confusion Matrix on TEST data - Actual vs. Predicted

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf, display_labels=model.classes_)
cfd.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

### Conclusion --> Above model is good becuase f1_macro score is more than 0.80

## Step 8: Out of Sample Predictions

```python
xnew = pd.read_csv('iris_sample.csv')
xnew
```

### 8(i): Apply pre.transfrom on xnew

```python
num_pipe
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

### 8(ii): Get the Predictions

```python
preds = model.predict(xnew_pre)
preds
```

```python
model.classes_
```

### 8(iii): Get the maximum probability of prediction in each class

```python
probs = model.predict_proba(xnew_pre)
probs
```

```python
import numpy as np
max_probs = np.max(probs, axis=1)
max_probs
```

### 8(iv): Save above results in DataFrame

```python
xnew['species_pred'] = preds
xnew['prob'] = max_probs
```

```python
xnew
```

```python
xnew['species_pred'].value_counts()
```

```python
xnew['species_pred'].value_counts().plot(kind='bar')
```

### 8(v): Save above DF to csv

```python
xnew.to_csv('irisResults.csv', index=False)
```
