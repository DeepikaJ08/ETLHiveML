## Step 1: Read the Dataset

```python
import pandas as pd
df = pd.read_csv('Cars93.csv', na_values=['', 'NA'], keep_default_na=False)
df.head()
```

## Step 2: Perform Basic Data Quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m[m>0]
```

```python
df.duplicated().sum()
```

## Step 3: Separate X and Y (Weight)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

## Step 4: Separate Cat & Con Features for X

```python
X.columns
```

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

## Step 6: Apply Train-Test Split

20% Test

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=60)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

## Step 7: Create Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

## Step 7: Build Lasso Model 

```python
from sklearn.linear_model import Lasso
model2 = Lasso(alpha=5)
model2.fit(xtrain, ytrain)
```

```python
model2.score(xtrain, ytrain)
```

```python
model2.score(xtest, ytest)
```

## Step 9: Hyper Parameter Tuning with GridSearchCV --> Aplha Tuning

```python
import numpy as np
params = {'alpha':np.arange(start=0.1, stop=100, step=0.1)}
params
```

```python
from sklearn.model_selection import GridSearchCV
ls = Lasso()
gscv = GridSearchCV(estimator=ls, param_grid=params, cv=5, scoring='neg_mean_squared_error')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_lasso = gscv.best_estimator_
best_lasso
```

## Step 10: Evaluate Lasso Model

```python
best_lasso.score(xtrain, ytrain)
```

```python
best_lasso.score(xtest, ytest)
```

### Step 10(i):  Cross Validate R2 score on Train

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(best_lasso, xtrain, ytrain, cv=5, scoring='r2')
```

```python
scores.mean()
```

#### You can using different scoring method but 'r2' in cross validation and 'neg_mean_squared_error' performs better in Regression

```python
scores2 = cross_val_score(best_lasso, xtrain, ytrain, cv=5, scoring='neg_mean_absolute_error')
scores2
```

```python
scores2.mean()
```
